## 1 — conda packages

```python
# ffmpeg via conda-forge (needed for video cutting)
# Note: this downgrades some Qt packages — that is fine, it does not affect this project
import subprocess
result = subprocess.run(
    ["conda", "install", "-c", "conda-forge", "ffmpeg", "-y"],
    capture_output=True, text=True
)
print(result.stdout[-2000:] if result.stdout else "")
print(result.stderr[-500:]  if result.stderr else "")
print("Return code:", result.returncode)
```

## 2 — pip packages

```python
import subprocess, sys
packages = [
    "numpy",
    "scipy",
    "matplotlib",
    "librosa",
    "imageio-ffmpeg",   # self-contained ffmpeg binary, avoids DLL issues on Windows
    "soundfile",        # audio I/O backend for librosa
    "scikit-learn",     # used in build_templates.py clustering
]
for pkg in packages:
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", pkg],
        capture_output=True, text=True
    )
    status = "OK" if result.returncode == 0 else "FAILED"
    print(f"  [{status}]  {pkg}")
    if result.returncode \!= 0:
        print(result.stderr[-300:])
```

## 3 — Verify everything imports correctly

```python
import importlib, shutil

checks = {
    "numpy":           "numpy",
    "scipy":           "scipy",
    "matplotlib":      "matplotlib",
    "librosa":         "librosa",
    "soundfile":       "soundfile",
    "imageio_ffmpeg":  "imageio_ffmpeg",
    "sklearn":         "scikit-learn",
}

all_ok = True
for module, label in checks.items():
    try:
        importlib.import_module(module)
        print(f"  [OK]      {label}")
    except ImportError:
        print(f"  [MISSING] {label}  <-- re-run the pip cell above")
        all_ok = False

# Check ffmpeg
try:
    import imageio_ffmpeg
    ff = imageio_ffmpeg.get_ffmpeg_exe()
    print(f"  [OK]      ffmpeg (imageio-ffmpeg)  ->  {ff}")
except Exception as e:
    print(f"  [MISSING] ffmpeg: {e}")
    all_ok = False

print()
if all_ok:
    print("All good\! Restart the kernel, then open 01_make_params.ipynb.")
else:
    print("Some packages are missing. Re-run the cells above, then try again.")
```

## 4 — Restart kernel

After all checks pass, do **Kernel → Restart Kernel** (or the restart button at the top).
This ensures the newly installed packages are fully loaded before you run the other notebooks.
